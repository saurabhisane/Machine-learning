# Visualizing missing data

```python
sns.heatmap(df.isnull(), cmap='viridis', cbar=False, yticklabels=False)
plt.title("Missing Data Heatmap")
plt.show()
```

```python
# Handling missing data
# Fill Age with median, Embarked with mode, Drop Cabin (too many missing values)


df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df.drop(columns=['Cabin'], inplace=True)
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Survived', hue='Sex', data=df)
plt.title("Survival Count by Gender")
plt.show()
```

# Correlation matrix

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.select_dtypes(include=['number']).corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Feature Correlation Matrix")
plt.show()
```

First-class passengers and those who paid higher fares had a better chance of survival, while third-class passengers had the lowest survival rate. Younger passengers and those traveling with family had slightly better chances. The strongest predictor of survival is ticket class (Pclass), followed by Fare, while Age and family size have a smaller impact.
